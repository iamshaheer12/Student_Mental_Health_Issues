```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# Load your dataset
# Replace 'your_dataset.csv' with the actual path to your dataset
data = pd.read_csv('my dataset.csv')

# Assuming the 'mental_health_issue' column is the target variable
X = data.drop('mental_health_issue', axis=1)
y = data['mental_health_issue']

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# List of models to try
models = [
    ('Random Forest', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('SVM', SVC()),
    ('Logistic Regression', LogisticRegression()),
    ('KNN', KNeighborsClassifier())
]

# Evaluate models using cross-validation
for name, model in models:
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    print(f'{name}: Mean Accuracy - {np.mean(scores):.2%}, Standard Deviation - {np.std(scores):.2%}')

# Choose the best model based on cross-validation results and evaluate on the test set
best_model = RandomForestClassifier(n_estimators=100, random_state=42)  # Replace with the best-performing model
best_model.fit(X_train_scaled, y_train)
test_accuracy = best_model.score(X_test_scaled, y_test)
print(f'\nBest Model Test Accuracy: {test_accuracy:.2%}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.manifold import TSNE

# Load the dataset
data = pd.read_csv('my dataset.csv')

# Add labels to mental health issues
mental_health_labels = {
    0: 'No mental health issue',
    1: 'Depression',
    2: 'Anxiety',
    3: 'Stress',
    4: 'Other mental health issues'
}
data['mental_health_issue'] = data['mental_health_issue'].map(mental_health_labels)

# Extract features and target variable
X = data.drop('mental_health_issue', axis=1)
y = data['mental_health_issue']

# Encode mental health labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a RandomForestClassifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Inverse transform the predictions to get original mental health labels
y_pred_labels = label_encoder.inverse_transform(y_pred)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Model Accuracy: {accuracy:.2%}')

# Display classification report and confusion matrix
print('\nClassification Report:')
print(classification_report(y_test, y_pred))

# Display confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('\nConfusion Matrix:')
print(conf_matrix)

# Visualize data using t-SNE (2D scatter plot)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot based on mental health issues
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_encoded, cmap='viridis', alpha=0.7)
plt.title('t-SNE Visualization based on Mental Health Issues')
plt.legend(*scatter.legend_elements(), title='Mental Health Issue')

# Add a test case: Visualize data based on gender
plt.figure(figsize=(12, 8))
scatter_gender = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=data['gender'], cmap='coolwarm', alpha=0.7)
plt.title('t-SNE Visualization based on Gender')
plt.legend(*scatter_gender.legend_elements(), title='Gender')

# Add a test case: Visualize data based on age
plt.figure(figsize=(12, 8))
scatter_age = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=data['age'], cmap='plasma', alpha=0.7)
plt.title('t-SNE Visualization based on Age')
plt.colorbar(label='Age')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.manifold import TSNE

# Load the dataset
data = pd.read_csv('my dataset.csv')

# Add labels to mental health issues
mental_health_labels = {
    0: 'No mental health issue',
    1: 'Depression',
    2: 'Anxiety',
    3: 'Stress',
    4: 'Other mental health issues'
}
data['mental_health_issue'] = data['mental_health_issue'].map(mental_health_labels)

# Select the first 8 features for visualization
selected_features = data.columns[9:17]

# Extract features and target variable
X = data[selected_features]
y = data['mental_health_issue']

# Encode mental health labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# One-hot encode categorical variables
X = pd.get_dummies(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a RandomForestClassifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Inverse transform the predictions to get original mental health labels
y_pred_labels = label_encoder.inverse_transform(y_pred)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Model Accuracy: {accuracy:.2%}')

# Display classification report and confusion matrix
print('\nClassification Report:')
print(classification_report(y_test, y_pred))

# Display confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('\nConfusion Matrix:')
print(conf_matrix)

# Visualize distribution of mental health issues based on the first 8 features
plt.figure(figsize=(14, 8))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(2, 4, i)
    sns.countplot(x=feature, hue='mental_health_issue', data=data)
    plt.title(f'Distribution based on {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.legend(title='Mental Health Issue')

plt.tight_layout()
plt.show()
```
